## Data preprocessing

### Subtask:
Prepare the data for machine learning by handling missing values, encoding categorical variables, scaling numerical features, and splitting the data into training and testing sets.


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np
import pandas as pd # Import pandas here to ensure it's available

# Reload the dataframe with the correct delimiter
# Assuming 'csv_file_name' and 'uploaded' are available from previous cells
# Get the actual uploaded file name from the uploaded dictionary
actual_zip_file_name = list(uploaded.keys())[0]
nested_zip_file_name = 'student.zip'
csv_file_name = None

with zipfile.ZipFile(io.BytesIO(uploaded[actual_zip_file_name]), 'r') as z:
     with z.open(nested_zip_file_name) as nested_z:
        with zipfile.ZipFile(io.BytesIO(nested_z.read()), 'r') as inner_z:
            for file_info in inner_z.infolist():
                if file_info.filename.endswith('.csv'):
                    csv_file_name = file_info.filename
                    break

if csv_file_name:
    df = pd.read_csv(csv_file_name, delimiter=';')
    print(f"'{csv_file_name}' loaded into DataFrame with semicolon delimiter.")
else:
    print("Error: CSV file not found.")
    # Exit if the CSV is not found to avoid further errors
    exit()


# Identify categorical and numerical columns
categorical_features = df.select_dtypes(include=['object']).columns
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Define the target variables
target_regression = 'G3'
target_classification = 'passed' # We will create this target variable later

# Remove the regression target variable from the numerical features list if it exists
if target_regression in numerical_features:
    numerical_features.remove(target_regression)

# Create transformers for numerical and categorical features
# Add an imputer for missing values (although we found none, it's good practice)
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Create a column transformer to apply different transformations to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Create the classification target variable (assuming pass if G3 >= 10)
df['passed'] = np.where(df['G3'] >= 10, 1, 0)

# Separate features and targets for both tasks
X = df.drop([target_regression, 'passed'], axis=1) # Drop both G3 and the temporary 'passed' column for features
y_regression = df[target_regression]
y_classification = df['passed']

# Split data into training and testing sets for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_regression, test_size=0.2, random_state=42)

# Split data into training and testing sets for classification
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X, y_classification, test_size=0.2, random_state=42, stratify=y_classification)


# Apply preprocessing to the training and testing sets for regression
X_train_reg_processed = preprocessor.fit_transform(X_train_reg)
X_test_reg_processed = preprocessor.transform(X_test_reg)

# Apply preprocessing to the training and testing sets for classification
X_train_clf_processed = preprocessor.fit_transform(X_train_clf)
X_test_clf_processed = preprocessor.transform(X_test_clf)


print("Data preprocessing and splitting complete for both regression and classification tasks.")
print("Shape of X_train_reg_processed:", X_train_reg_processed.shape)
print("Shape of X_test_reg_processed:", X_test_reg_processed.shape)
print("Shape of y_train_reg:", y_train_reg.shape)
print("Shape of y_test_reg:", y_test_reg.shape)
print("\nShape of X_train_clf_processed:", X_train_clf_processed.shape)
print("Shape of X_test_clf_processed:", X_test_clf_processed.shape)
print("Shape of y_train_clf:", y_train_clf.shape)
print("Shape of y_test_clf:", y_train_clf.shape) # Corrected typo here
```
